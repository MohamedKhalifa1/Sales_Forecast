```python
DATA_PATH = 'splited data/'
```

```python
import pandas as pd 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X_train = pd.read_pickle('{}/X_train.pkl'.format(DATA_PATH))
X_test = pd.read_pickle('{}/X_test.pkl'.format(DATA_PATH))
y_train = pd.read_pickle('{}/y_train.pkl'.format(DATA_PATH))
y_test = pd.read_pickle('{}/y_test.pkl'.format(DATA_PATH))
```

```python
linearRegressionModel = LinearRegression()
```

```python
linearRegressionModel.fit(X_train,y_train)
```

```python
y_pred = linearRegressionModel.predict(X_test)
```

```python
y_pred
```

```python
ac = r2_score(y_true=y_test,y_pred=y_pred)
print("model Accurcy = {}%".format(ac*100))
```

```python
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
```

```python
data = {'Actual': y_test, 'Predicted': y_pred}
y_comparison = pd.DataFrame(data)
y_comparison
```

```python
X_test_df = pd.DataFrame(X_test, columns=X_test.columns)
test_df = X_test_df.copy()
test_df['y_pred'] = y_pred
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

for i in test_df.columns:
    plt.figure(figsize=(8, 6))
    sns.regplot(x=i, y='y_pred', data=test_df, scatter_kws={'alpha':0.5})

    # Set labels
    plt.xlabel(i)
    plt.ylabel('Predicted Values')

    plt.show()
```
