```python
FILE_PATH = 'data\salesforcast_preprocessed.csv'
```

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv(FILE_PATH)
```

```python
df.head()
```

```python
df['Gross Profit'] = df['Revenue'] - df['Cost']
```

```python

df['Gross Margin'] = (df['Gross Profit'] / df['Revenue']) * 100
```

```python
df['Return on Investment (ROI)'] = (df['Gross Profit'] / df['Cost'])*100
```

```python
df['Unit Profit'] = df['Unit Price'] - df['Unit Cost']
```

```python
df.head()
```

```python
plt.figure(figsize=(8,10))
plt.hist(x='Unit Price',data=df)
plt.show()
```

```python
plt.figure(figsize=(8,10))
plt.hist(x='Unit Cost',data=df)
plt.show()
```

```python
plt.figure(figsize=(8,10))
plt.hist(x='Revenue',data=df)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(x=df['Month'], y=df['Revenue'])
plt.title('Monthly Revenue')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
sns.barplot(x=df['Country'], y=df['Revenue'])
plt.title('Country Revenue')
plt.xlabel('Country')
plt.ylabel('Revenue')
plt.show()
```

```python
ger_states = df[df['Country'] == 'Germany']

plt.figure(figsize=(8, 10))
sns.barplot(x='State', y='Revenue',data=ger_states)
plt.title('State Revenue in the Germany')
plt.xlabel('State')
plt.ylabel('Revenue')
plt.show()
```

```python
ger_country = df[df['State'] == 'Brandenburg']
plt.figure(figsize=(8, 10))
plt.pie(ger_country['Customer Gender'].value_counts(), labels=ger_country['Customer Gender'].value_counts().index, autopct='%1.1f%%', startangle=90)
plt.title('Customer Gender vs Revenue in the Brandenburg')
plt.show()
```
